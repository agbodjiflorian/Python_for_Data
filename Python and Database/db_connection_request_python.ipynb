```python
import psycopg2

def connexion_db():
    DB_NAME = "foodie_fi" #modifier
    DB_USER = "postgres" #modifier
    DB_PASS = "sodandjo1962" #modifier
    DB_HOST = "localhost" #modifier
    DB_PORT = "5432" #ne pas modifier, c'est le port PostgreSQL

    try:
    	conn = psycopg2.connect(database=DB_NAME,
							user=DB_USER,
							password=DB_PASS,
							host=DB_HOST,
							port=DB_PORT)
    	print("Database connectée")
    except:
    	print("Database non connectée")

    return (conn)
```

### 2- Créer une table

```python
def creer_table():
    conn = connexion_db()
    # Créer un cursor. Un objet de la classe cursor qui permet d'exécuter la requete SQL voulue
    cur = conn.cursor()

    # Elle prend une chaine de caractère qui correspond à la requete SQL voulue 
    cur.execute("DROP TABLE IF EXISTS test")

    cur.execute("""
        CREATE TABLE test (
            id SERIAL PRIMARY KEY,
            nom TEXT,
            prenom TEXT,
            age INTEGER,
            classe TEXT
        )
    """)
    
    # Sauvegarder la requete
    conn.commit()

    # Fermer la connexion
    #conn.close()
    return (conn)
```

### 3- Insérer des éléménts dans une table
