# Clean the data

```python
# remove all the rows where the date is unknown
df_accident = df_accident[df_accident['date'] != 'date unk.']
df_accident
```

```python
df_accident_clean = df_accident
```

```python
# check the date format properly
df_accident_clean['date'] = df_accident_clean['date'].str.strip()
df_accident_clean
```

```python
df_accident_clean
```

```python
df_accident_clean.info()
```

```python
df_accident_new = df_accident_clean
```

```python
df_accident_new
```

```python
# split the date column in three other columns to clean the data column based on the errors in each new columns(day, month, year)

df_accident_new[['Day', 'Month', 'Yearr']] = df_accident_new.date.str.split("-", expand = True)

df_accident_new
```

```python
df_accident_new.nunique()
```

```python
# add a boolean value True to all day columns containing a digit
df_accident_new['Day Number'] = df_accident_new['Day'].str.isdigit() 
```

```python
df_accident_new
```

```python
df_accident_new.group_by(['Day Number']).count()
```

```python
# select all the rows where the Day isn't a digit
df_accident_new[~df_accident_new['Day Number']]
```

```python
# replace all the NaN values with an empty string
df_accident_new = df_accident_new.fillna('')
df_accident_new
```

```python
df_accident_new[~df_accident_new['Day Number']]
```

```python
# replace the empty values with 0
df_accident_new['fatalities'] = df_accident_new['fatalities'].replace('', 0)
df_accident_new
```

```python
# replace the empty values with unknown
df_accident_new['location'] = df_accident_new['location'].replace('', 'unknown')
df_accident_new
```

```python
df_accident_new[~df_accident_new['Day Number']]
```

```python
# drop all the rows with an empty Month
df_accident_new = df_accident_new.drop(df_accident_new[df_accident_new['Month'] == "???"].index)
```

```python
# after dropping all the rows with an empty month, I created a new date column by giving a specific value (1st of the month, 01) to the rows with an empty day

df_accident_new['new_date'] = df_accident_new['Day'] + '-' + df_accident_new['Month'] + '-' + df_accident_new['Yearr']
df_accident_new
```

```python
from datetime import datetime
```

```python
# change the months notation from capital letters to numbers 
months = {"JAN" : "01",
          "FEB" : "02",
          "MAR" : "03",
          "APR" : "04",
          "MAY" : "05",
          "JUN" : "06",
          "JUL" : "07",
          "AUG" : "08",
          "SEP" : "09",
          "OCT" : "10",
          "NOV" : "11",
          "DEC" : "12"
         }

for key, value in months.items():
    df_accident_new['Month'] = df_accident_new['Month'].replace(key, value)
    
```

```python
df_accident_new['new_date'] = df_accident_new['Day'] + '-' + df_accident_new['Month'] + '-' + df_accident_new['Yearr']
df_accident_new
```

```python
df_accident_new['new_date'] = df_accident_new['new_date'].str.replace('-', '/')
df_accident_new
```

```python
# change the date format from str to datetime

def subtract_3_days(match):
    days = int(match.group(1))
    return str(days - 3)

tempdate = pd.to_datetime(df_accident_new["new_date"], format="%d/%m/%Y", errors="coerce")
fixed_dates = df_accident_new.loc[tempdate.isna(), "new_date"].str.replace(r"(?<=/)(\d+)(?=/)", subtract_3_days)
tempdate.update(pd.to_datetime(fixed_dates, format="%d/%m/%Y"))

df_accident_new["clean_date"] = tempdate
df_accident_new
```

```python
df_accident_new.dtypes
```

```python
# get the week days from 0 (monday) to 6 (sunday)
df_accident_new['day_of_week'] = df_accident_new['clean_date'].dt.dayofweek
df_accident_new
```

```python
# add a column with the week day names
days = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

df_accident_new['day_of_week'] = df_accident_new['day_of_week'].map(days)

df_accident_new
```

```python
# delete the old date column
df_accident_new = df_accident_new.drop(columns = ['date', 'Day Number', 'Yearr', 'new_date'])
```

```python
df_accident_new.dtypes
```

```python
df_accident_new = df_accident_new.rename(columns={'clean_date': 'date', 'Month': 'month', 'Day':'day', 'nMonth':'month_name'}) 
df_accident_new 
```

```python
# set the date column as index

df_accident_new.set_index('date', inplace=True)
```

```python
# number of incident per month per country
df3 = pd.DataFrame(df_accident_new.groupby('country')['month_name'].value_counts())
df3 = df3.rename(columns={'month_name':'nb_incidents'})
df3
```
